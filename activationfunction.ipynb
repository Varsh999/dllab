```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Activation Functions
# -----------------------------

def threshold(x):
    return np.where(x >= 0, 1, 0)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# -----------------------------
# Input Values (small for output)
# -----------------------------
x_small = np.array([-2, -1, 0, 1, 2])

# Apply functions
print("Input:", x_small)

print("\nThreshold Output:")
print(threshold(x_small))

print("\nSigmoid Output:")
print(sigmoid(x_small))

print("\nReLU Output:")
print(relu(x_small))

print("\nTanh Output:")
print(tanh(x_small))


# -----------------------------
# Graph Part (large values)
# -----------------------------
x = np.linspace(-10, 10, 200)

plt.figure()

plt.plot(x, threshold(x), label="Threshold")
plt.plot(x, sigmoid(x), label="Sigmoid")
plt.plot(x, relu(x), label="ReLU")
plt.plot(x, tanh(x), label="Tanh")

plt.title("Activation Functions")
plt.xlabel("Input (x)")
plt.ylabel("Output")
plt.legend()
plt.grid()

plt.show()
```
